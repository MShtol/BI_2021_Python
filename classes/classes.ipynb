**1. Create class with constructor and couple of methods**

```python
class VideoGame:
    def __init__(self, name, genre='action'):
        self.name = name
        self.genre = genre
        self.developer = 'Nonanme'
        self.dimensions = 3
        self.character = ['wood elves','castle guard','villain']
        self.villians = ['castle guard', 'villain']
        self.caravan = False
        self.ideas = []
    

    def set_caravan(self):
        """
        Adds caravan robing
        """
        self.caravan = True
        
    def set_dev(self, name):
        """
        Sets developers name
        """
        self.developer = name
        
    def set_dims(self, n):
        """
        Sets number of dimensuins in game
        """
        self.dimensions = n
        
    def set_char(self, *args):
        """
        Sets one or list of playable characters
        """
        self.character = args
        
    def set_vill(self, *args):
        """
        Sets one or list of ugly villians
        """
        self.villians = args
        
    def add_idea(self, idea):
        """
        Adds ideas for implementation
        """
        self.ideas += [idea]
        
    def evaluation(self):
        if self.caravan:
            return "You can make bilions"
        else:
            return "You can't sell it. Consider caravan robing"
        
    def concept(self):
        """
        Prints a message that contains full concept for investors
        """
        print(f'''Hello! My name is {self.developer}.
I want to create a game in {self.dimensions}-D-{self.genre} genre! It's called {self.name}.
User can play as {", ".join(self.character)}.
And if player is playing as {self.character[0]} then he protects wooden houses from {", ".join(self.villians)}.''')
        if self.caravan:
            print('''You can robe caravans here!!!
            ''')
        else:
            print("""You can not robe any caravans :( It's a safe place
            """)
        print(f'''Other ideas:''')
        for idea in self.ideas:
            print(idea)
        print('_________________________________________')
```

```python
ds = VideoGame('Death stranding', "walking-simulator")
ds.set_caravan()
ds.set_dev('Hideo Kojima. Am I genius?')
ds.set_dims(3.5)
ds.set_char('Norman Reedus')
ds.set_vill("heavy rain","raiders")
ds.add_idea("You have a canister with pet baby")
ds.add_idea('Add some Apocalyptica melody')
ds.add_idea('Indirect multyplayer, maybe')
ds.concept()
print(ds.evaluation())
```

**2. Create class RNA with transcription and translation methods**

```python
from Bio.Seq import translate
from Bio.Seq import back_transcribe


class RNA:
    code = {'UUU': 'F', 'UUC': 'F', 'UUA': 'L', 'UUG': 'L', 'UCU': 'S', 'UCC': 'S', 'UCA': 'S',
                   'UCG': 'S', 'UAU': 'Y', 'UAC': 'Y', 'UAA': '*', 'UAG': '*', 'UGU': 'C', 'UGC': 'C',
                   'UGA': '*', 'UGG': 'W', 'CUU': 'L', 'CUC': 'L', 'CUA': 'L', 'CUG': 'L', 'CCU': 'P',
                   'CCC': 'P', 'CCA': 'P', 'CCG': 'P', 'CAU': 'H', 'CAC': 'H', 'CAA': 'Q', 'CAG': 'Q',
                   'CGU': 'R', 'CGC': 'R', 'CGA': 'R', 'CGG': 'R', 'AUU': 'I', 'AUC': 'I', 'AUA': 'I',
                   'AUG': 'M', 'ACU': 'T', 'ACC': 'T', 'ACA': 'T', 'ACG': 'T', 'AAU': 'N', 'AAC': 'N',
                   'AAA': 'K', 'AAG': 'K', 'AGU': 'S', 'AGC': 'S', 'AGA': 'R', 'AGG': 'R', 'GUU': 'V',
                   'GUC': 'V', 'GUA': 'V', 'GUG': 'V', 'GCU': 'A', 'GCC': 'A', 'GCA': 'A', 'GCG': 'A',
                   'GAU': 'D', 'GAC': 'D', 'GAA': 'E', 'GAG': 'E', 'GGU': 'G', 'GGC': 'G', 'GGA': 'G',
                   'GGG': 'G'}
    rna_rev_dna = {'C':'G', "G":"C","A":'T', "U":"A"}
    
    def __init__(self, seq):
        self.wrong_len = False
        self.raw_seq = seq.upper()
        if len(seq)%3:
            self.wrong_len = True
            self.seq = seq[:-(len(seq)%3)]
            print("sequence has unended codon. Translation wull be performed with trimmed sequence")
        else:
            self.seq = self.raw_seq
            
    
    def translate(self):
        """
        Translates RNA sequence. If it has unended codons, will be translated trimmed sequence
        """
        protein = []
        for i in range(0,len(self.seq),3):
            protein += self.code[self.seq[i:i+3]] 
        return ''.join(protein)
    
    
    def rev_transcript(self):
        """
        Reverse transcription of sequence
        """
        dna = []
        for i in range(len(self.raw_seq)):
            dna += self.rna_rev_dna[self.raw_seq[i]]
        return ''.join(dna)
```

**3. Class derived from set()**


```python
class PosSet(set):
    
    
    def __init__(self, numbers):
        self.numbers = [number for number in numbers if isinstance(number, (int,float)) and number > 0]
        super().__init__(self.numbers)
    
    def add(self, number):
        """
        Adds positive number to set.
        If element is something else, it's discarded.
        """
        if isinstance(number, (int,float)) and (number > 0):
            super().add(number)
        
    
```

**4.Create Class for stats calculation of fasata files**


```python
class StatoFasta:
    
    def __init__(self, path):
        self.path = path
        self.lens = None
        self.gc_content = None
        self.count = None
        self.nuc = None
        
        
    def count_seq(self):
        """
        Counts the number of sequences in the FASTA file
        """
        with open(self.path) as f:
            self.count = f.read().count('>')
            return self.count
            
    def gc_count(self):
        """
        Calculates GC content for every sequence in fasta.
        Return list of values
        """
        self.gc_content = []
        with open(self.path) as f:
            for seq in SeqIO.parse(f, "fasta"):
                self.gc_content.append(np.round(GC(seq.seq), 2))

        return self.gc_content
        
    
    def len_calc(self):
        """
        Calculates length of every sequence.
        Returns list of lengths.
        """
        self.lens = []
        with open(self.path) as f:
            for seq in SeqIO.parse(f, "fasta"):
                self.lens.append(len(seq.seq))

        return self.lens
    
    
    def plot_len_distr(self):
        """
        Plot histogram with distribution of sequences' lenghts.
        """
        if self.lens is None:
            self.len_calc()
            
        sns.histplot(self.lens)
        plt.xlabel("Sequence length,base pairs")
        plt.title("Length disribution of sequences")
        plt.show()
        
    
    def which_nuc(self):
        dna_nucl = "ACGT"
        rna_nucl = "ACGU"
        nucleoset = set()
        
        with open(self.path) as f:
                for seq in SeqIO.parse(f, "fasta"):
                    nucleoset.update([sequence.upper() for sequence in seq.seq])
                                     
        if nucleoset.issubset(set(dna_nucl)):
            self.nuc = "DNA"
            self.tetra = {a+b+c+d:0 for a in dna_nucl for b in dna_nucl for c in dna_nucl for d in dna_nucl}
        elif nucleoset.issubset(set(rna_nucl)):
            self.nuc = "RNA"
            self.tetra = {a+b+c+d:0 for a in rna_nucl for b in rna_nucl for c in rna_nucl for d in rna_nucl}
        else:
            self.nuc = "non-nucleoacide"

  
    def tetra_distr(self):
        """
        Calculates and plots tetramer distribution
        """
        if self.nuc is None:
            self.which_nuc()
        if self.nuc == "non-nucleoacide":
            print("tetramer_distr will not be calculated. Sequence is not nucleo")
        else:
            with open(self.path) as f:
                for seq in SeqIO.parse(f, "fasta"):
                    for i in range(len(seq.seq) - 3):
                        self.tetra[str(seq.seq[i:i + 4]).upper()] += 1
        
        total_tetra = sum(self.tetra.values())
        for key,value in self.tetra.items():
            self.tetra[key] = value/total_tetra

        plt.figure(figsize=(30, 10))
        plt.bar(self.tetra.keys(), self.tetra.values())
        plt.xlabel("Tetramer")
        plt.ylabel("Frequency, %")
        plt.xticks(rotation=90, fontsize=5)
        plt.title("Distribution of tetramers across sequnces", fontsize=14)   
        plt.show()
        
        
    def __str__(self):
        """
        Redefines what str() returns for object (and subsequentlu - print())
        """
        return f"path to file is {self.path}"
    
    
    def calc_stat(self):
        """
        Calculates and print implemented statistics
        """
        self.len_calc()
        self.count_seq()
        self.gc_count()
        self.which_nuc()
        self.tetra_distr()
        print(f"""There're {self.count} sequences of {self.nuc} in file. 
Their lengths are: 
{self.lens}
Their GC-contents are:
{self.gc_content}
Average tetramer content across sequences was calculated.
""")
        
```

**5. Abstract classes** 
(Done together with https://github.com/AnasZol)

Write an abstract class `BiologicalSequence` that defines the following interface:
+ Working with the `len` function
+ Ability to get elements by index and slice sequences (similar to strings)
+ Printing in a convenient form and the ability to convert to a string
+ Ability to check the alphabet of the sequence for correctness

Write a `NucleicAcidSequence` class:
+ This class implements the `BiologicalSequence` interface
+ This class has a new `complement` method that returns a complementary sequence
+ This class has a new `gc_content` method that returns the GC content (no difference, percentage or fraction)

Write `NucleicAcidSequence` descendant classes: `DNASequence` and `RNASequence`
+ `DNASequence` must have a `transcribe` method that returns the transcribed RNA sequence
+ These classes should not have <ins>public methods</ins> `complement` and a method for checking the alphabet, since they must already be implemented in `NucleicAcidSequence`.

Write an `AminoAcidSequence` class:
+ This class implements the `BiologicalSequence` interface
+ Add any one method to this class that matches the meaning of the amino acid sequence. For example, a method for finding the isoelectric point, molecular weight, etc.

```python
from abc import ABC, abstractmethod


class BiologicalSequence(ABC):
    '''An abstract class for biological sequences
    
    Abstract methods:
        __init__(): class constructor
        __getitem__(): gets item by slice or index
        __len__(): returns sequence length
        __str__(): makes readable sequence version
        __repr__(): for class representation
        check_alphabet(): checks if sequence correspondes to class alphabet
    '''
    
    @abstractmethod
    def __init__(self, seq):
        pass
     
        
    @abstractmethod
    def __getitem__(self, els):
        pass
    
    
    @abstractmethod
    def __len__(self):
        pass
    
    
    @abstractmethod
    def __str__(self):
        pass
    
    
    @abstractmethod
    def __repr__(self):
        pass
    
    
    @abstractmethod
    def check_alphabet(self):
        pass
    

class AminoNucSequence(BiologicalSequence):
    '''A class for implementation of BiologicalSequence interface'''
    
    def __init__(self, seq):
        if isinstance(seq, str):
            self.seq = seq.upper()
        else:
            raise NotImplementedError('NotImplemented')
        
     
    def __getitem__(self, els):
        return self.seq[els]
    
    
    def __len__(self):
        return len(self.seq)
    
    
    def __str__(self):
        return self.seq
    
    
    def __repr__(self):
        return f'{type(self).__name__}:\n{self.seq}'
    
    
    def check_alphabet(self):
        for letter in self.seq:
            if letter not in self.alphabet:
                return False
        return True
    
    
class NucleicAcidSequence(AminoNucSequence):
    '''A class for nucleic acid sequences
    
    Methods:
        complement(): makes coplement for nucleic acid
        gc_content(): counts GC-content of nucleic acid
    '''
    
    
    def complement(self):
        '''To complement nucleic acid sequence
        
        Returns: str, complemented nucleic acid sequence    
        '''
        
        
        assert self.check_alphabet(), 'Wrong alphabet'
        if hasattr(self, 'alphabet'):
             return self.seq.translate(str.maketrans(self.alphabet))
        else:
            raise NotImplementedError('NotImplemented')
    
    
    def gc_content(self):
        '''To calculate GC-content of nucleic acid sequence
        
        Returns: float, GC-share in nucleic acid sequence
        '''
        
        
        assert self.check_alphabet(), 'Wrong alphabet'
        return (self.seq.count('G') + self.seq.count('C')) / len(self.seq)

    
class DNASequence(NucleicAcidSequence):
    '''A class for DNA sequences
    
    Attributes:
        alphabet (dict): contains complement DNA nucleotides pairs
        
    Methods:
        transcribe(): transcribes nucleic acid sequence
    '''
    
    
    alphabet = {'A': 'T', 'T': 'A',
                'G': 'C', 'C': 'G'}
    
    def transcribe(self):
        '''To transcribe nucleic acid sequence
        
        Returns: str, transcribed nucleic acid sequence
        '''
        
        
        assert self.check_alphabet(), 'Wrong alphabet'
        return self.seq.replace('T', 'U')


class RNASequence(NucleicAcidSequence):
    '''A class for RNA sequences
    
    Attributes:
        alphabet (dict): contains complement RNA nucleotides pairs
    '''
    
    
    alphabet = {'A': 'U', 'U': 'A',
                'G': 'C', 'C': 'G'}

    
class AminoAcidSequence(AminoNucSequence):
    '''A class for amino acid sequences
    
    Attributes:
        alphabet (frozen): contains all amino acids
        
    Methods:
        extinction_coef(): calculates extinction coefficient for amino acid sequence
    '''

    
    alphabet = frozenset('GPAVLIMCFYWHKRQNEDST')
    
    
    def extinction_coef(self):
        '''Calculates extinction coefficient, 
        asuming all Cys (C) is in disulfide bonds
        
        Returns: int, calculated extinction coefficient
        '''
        
        
        assert self.check_alphabet, 'Wrong alphabet'
        n_W = self.seq.count('W')
        n_Y = self.seq.count('Y')
        n_C = self.seq.count('C')
        return n_W * 5500 + n_Y * 1490 + n_C * 125
    

if __name__ == '__main__':
    seq1 = DNASequence('ACGTACGTACGT')
    seq2 = RNASequence('ACGUACGUACGU')
    seq3 = AminoAcidSequence('GPAVLIMCFYWHKRQNEDST')
    seq4  = AminoAcidSequence('ZZZGPAVLIMCFYWZZZZHKRQNEDSTZZZ')
    print(seq1)
    print(seq1[:5], seq1[0])
    print(seq1.complement())
    print(seq1.transcribe())
    print(seq2.complement())
    print(seq3[1:])
    print(seq3.extinction_coef())
    print(seq3.check_alphabet(), seq4.check_alphabet())
```
